```python
#realise the functions sint,cost,sinht,and cosht for the vector t=[0,10] with the increment 0.01
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0,10,0.01)
x1=np.sin(t)
x2=np.cos(t)
x3=np.sinh(t)
x4=np.cosh(t)
fig=plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(t,x1)
plt.title("Sine function")
plt.xlabel("TIME")
plt.ylabel("AMPLITUDE")
plt.subplot(2,2,2)
plt.plot(t,x2)
plt.title("Cosine function")
plt.xlabel("TIME")
plt.ylabel("AMPLITUDE")
plt.subplot(2,2,3)
plt.plot(t,x3)
plt.title("hyperbolic sine function")
plt.xlabel("TIME")
plt.ylabel("AMPLITUDE")
plt.subplot(2,2,4)
plt.plot(t,x4)
plt.title("Hyperbolic cosine function")
plt.xlabel("TIME")
plt.ylabel("AMPLITUDE")
plt.show()

```
