```python
#first and second derivatives of functions
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,10,0.01)
x=np.sin(t)
y=np.cos(t)
a=np.sinh(t)
b=np.cosh(t)
def deri(m,n):
    d1=np.gradient(m,n)
    d2=np.gradient(d1,n)
    return d1,d2
d1x,d2x=deri(x,0.01)
d1y,d2y=deri(y,0.01)
d1a,d2a=deri(a,0.01)
d1b,d2b=deri(b,0.01)
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(t,x)
plt.plot(t,d1x)
plt.plot(t,d2x)
plt.title("sin and derivative")
plt.xlabel("values of t")
plt.ylabel("values of x")
plt.legend(["sin","first derivative","second derivative"],loc="lower right")
plt.grid()
plt.subplot(2,2,2)
plt.plot(t,y)
plt.plot(t,d1y)
plt.plot(t,d2y)
plt.title("cos and derivative")
plt.xlabel("values of t")
plt.ylabel("values of y")
plt.legend(["cos","first derivative","second derivative"],loc="lower right")
plt.grid()
plt.subplot(2,2,3)
plt.plot(t,a)
plt.plot(t,d1a)
plt.plot(t,d2a)
plt.title("sinh and derivative")
plt.xlabel("values of t")
plt.ylabel("values of a")
plt.legend(["sinh","first derivative","second derivative"],loc="upper left")
plt.grid()
plt.subplot(2,2,4)
plt.plot(t,b)
plt.plot(t,d1b)
plt.plot(t,d2b)
plt.title("cosh and derivative")
plt.xlabel("values of t")
plt.ylabel("values of b")
plt.legend(["cosh","first derivative","second derivative"],loc="upper left")
plt.grid()




```
